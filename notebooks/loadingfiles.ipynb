```python
import pickle
import faiss
import numpy as np

# Load KMeans clustering model
with open("models/kmeans.pkl", "rb") as f:
    kmeans = pickle.load(f)

# Load metadata for FAISS vectors
with open("models/meta.pkl", "rb") as f:
    meta = pickle.load(f)

# Load FAISS index
faiss_index = faiss.read_index("models/faiss_index.bin")

# Optional: Load image embeddings
try:
    with open("models/image_embeddings.pkl", "rb") as f:
        image_embeddings = pickle.load(f)
    print(f"✅ Loaded image embeddings: {len(image_embeddings)} items")
except FileNotFoundError:
    image_embeddings = None
    print("⚠️ image_embeddings.pkl not found, skipping")

print("✅ Loaded models and FAISS index")
```

```python
print(type(meta))
print(meta[:5])  # print first 5 entries to inspect
```

```python
# Suppose we have a query embedding (replace with actual embedding)
query_vector = np.random.rand(faiss_index.d).astype("float32").reshape(1, -1)

# Search top 5 similar vectors
k = 5
distances, indices = faiss_index.search(query_vector, k)

# Display results
print("✅ Top 5 similar items:")
for idx, dist in zip(indices[0], distances[0]):
    item_id = meta[idx]  # meta is a list of strings (IDs)
    print(f"ID: {item_id}, Distance: {dist}")
```

```python
# Suppose we have a query embedding (replace with actual embedding)
query_vector = np.random.rand(faiss_index.d).astype("float32").reshape(1, -1)

# Search top 5 similar vectors
k = 5
distances, indices = faiss_index.search(query_vector, k)

# Display results
print("✅ Top 5 similar items:")
for idx, dist in zip(indices[0], distances[0]):
    item_id = meta[idx]  # meta is a list of uniq_id strings
    # Map to cleaned_df to get title
    item_info = cleaned_df[cleaned_df['uniq_id'] == item_id].iloc[0]
    print(f"ID: {item_id}, Name: {item_info['title']}, Distance: {dist}")
```

```python
if image_embeddings is not None:
    print("\n✅ Corresponding image embeddings for top items:")
    for idx in indices[0]:
        img_emb = image_embeddings[idx]
        print(f"ID: {meta[idx]}, Image embedding shape: {img_emb.shape if hasattr(img_emb, 'shape') else 'N/A'}")

```
